```python
#1. Days to go for a particular date in future.
from datetime import date

to = date.today()
dt = input('Enter date as yyyy-mm-dd')
year, month, day = (int(x) for x in dt.split('-'))  
fd = date(year, month, day)

time_to_future = abs(fd - to).days
print(f'{time_to_future} more days to go..')
```

```python
#5. A dict contains events and corresponding dates, given a date, find out what events have occurred and what are still pending.
from datetime import date
events = { 'a' : '2022-01-01',
           'b' : '2022-03-23',
           'c' : '2022-04-17',
           'd' : '2022-04-22'
         }
dt = input('Enter date as yyyy-mm-dd ')
occured = []
pending = []
fd=datetime.fromisoformat(dt)
for k,v in events.items():
    li= datetime.fromisoformat(v)
    if li < fd :
        occured.append(k)
    else:
        pending.append(k)
print(f'Pending events : {pending}')
print(f'Occured events : {occured}')
```

```python
#7 Given a file with strings, calculate the permutations.
with open('../Downloads/permutations.txt') as file:
    words = [(line.strip()) for line in file]
n = len(words)
print(n)
def permutation(num):
    factorial = 1

    if num < 0:
        print("Empty file")
    elif num == 0:
        print("Empty file")
    else:
        for i in range(1,num + 1):
            factorial = factorial*i
    return factorial    
permutation(n)
    
```

```python
#12 Calculate how many times a string occurs in a given file.
with open('../Downloads/repeat.txt') as file:
    words = [(line.strip()) for line in file]
words_freq={}    

for i in range(len(words)):
    if(words[i] not in words_freq):
        words_freq[words[i]] = 1
    else:
        words_freq[words[i]]+=1
print(words_freq)  
```

```python
#15 Print all the files in a given directory.
import os
import time

ls = os.listdir('../Downloads/')

path = '../Downloads/'

for file in ls:
    file_path = f'{path}\\{file}'
    if os.path.isfile(file_path):
        try:
            # Printing the file pertaining to file_path
            os.startfile(file_path, 'print')
            print(f'Printing {file}')
            time.sleep(5)
        except:
            # Catching if any error occurs and alerting the user
            print(f'ALERT: {file} could not be printed! Please check\
            the associated softwares, or the file type.')
    else:
        print(f'ALERT: {file} is not a file, so can not be printed!')
         
print('Task finished!')
```
